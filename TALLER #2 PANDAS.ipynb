```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\anfre\Downloads\Listado-de-proveedores-y-contactos.csv",encoding="latin-1",sep=";")

df
```

```python
## Se elimina la columna Unnamed 7
df = df.drop('Unnamed: 7', axis=1)
```

```python
df
```

```python
## ¿Cuál es el proveedor al que se le realizó el pedido de mayor valor?

saldo_maximo = df['Saldo pendiente'].max()
print(saldo_maximo)
```

```python
fila_saldo_maximo = df.loc[df['Saldo pendiente'].idxmax()]
print(fila_saldo_maximo)
```

```python
# ¿Cuales son los provedores con un saldo superior a 8000?
proveedores_alta_deuda = df[df['Saldo pendiente'] > 8000]

# Mostrar el resultado
print(proveedores_alta_deuda)
```

```python
resultado_limpio = df.loc[df['Saldo pendiente'] > 8000, ['Proveedor', 'Saldo pendiente']]

# Mostrar el resultado
print(resultado_limpio)
```

```python
## ¿Cuál ha sido el valor total gastado en proveedores durante cada año?

# 2. Convertir 'Fecha de última compra' a tipo datetime
# Indicamos dayfirst=True ya que tus fechas están en formato día/mes/año (ej. 4/09/2018)
df['Fecha de última compra'] = pd.to_datetime(df['Fecha de última compra'], dayfirst=True)

# 3. Crear una nueva columna solo con el año
df['Año'] = df['Fecha de última compra'].dt.year

# 4. Agrupar por el año y calcular la suma del saldo
gasto_por_año = df.groupby('Año')['Saldo pendiente'].sum().reset_index()

# 5. Cambiar el nombre de las columnas para que sea más claro y mostrar el resultado
gasto_por_año.columns = ['Año', 'Total Gastado']
print(gasto_por_año)
```
