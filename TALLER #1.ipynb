## Estructuras de Datos (Listas, Tuplas, Diccionarios)
Crea un programa para organizar información de múltiples ventas utilizando las estructuras adecuadas según la naturaleza de los datos.

- Tupla: Crea una tupla con las categorías de productos permitidas en la tienda (ej. "Electrónica", "Ropa", "Alimentos"). Se usa una tupla porque las categorías de la tienda no cambian todos los días.
- Diccionario: Crea un diccionario que represente una venta específica. Debe tener las llaves: producto, categoria, precio, cantidad y total.
- Lista: Crea una lista vacía llamada registro_ventas y añade el diccionario de la venta que acabas de crear.

```python
## Haz tu codigo aquí:

# Tupla 
categorias = ("Electrónica", "Ropa", "Alimentos", "Farmacia")

# Solicitar datos de la venta
producto = input("Ingrese el nombre del producto: ")
categoria = input("Ingrese la categoría del producto: ")
precio = float(input("Ingrese el precio del producto: "))
cantidad = int(input("Ingrese la cantidad vendida: "))

# Calcular total
total = precio * cantidad

# Diccionario 
venta = {
    "producto": producto,
    "categoria": categoria,
    "precio": precio,
    "cantidad": cantidad,
    "total": total
}

# Lista para almacenar ventas
registro_ventas = []

# Añadir la venta
registro_ventas.append(venta)

print("\nCategorías permitidas:", categorias)
print("\nRegistro de ventas:")
print(registro_ventas)
```

## Estructuras de Control (Condicionales y Ciclos)
Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).

1. Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
2. Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
3. Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).

```python
## Haz tu codigo aquí:

registro_ventas = [
    {
        "producto": "Laptop",
        "categoria": "Electrónica",
        "precio": 2500000,
        "cantidad": 1,
        "total": 2500000
    },
    {
        "producto": "Camiseta",
        "categoria": "Ropa",
        "precio": 50000,
        "cantidad": 2,
        "total": 100000
    },
    {
        "producto": "Televisor",
        "categoria": "Electrónica",
        "precio": 1500000,
        "cantidad": 1,
        "total": 1500000
    }
]

for venta in registro_ventas:
    total_original = venta["total"]

    # Aplicar descuento
    if total_original > 100000:
        total_con_descuento = total_original * 0.90
    else:
        total_con_descuento = total_original

    print("\n--- Reporte de Venta ---")
    print("Producto:", venta["producto"])
    print("Total original: $", total_original)
    print("Total con descuento: $", total_con_descuento)
```

## Funciones y Manejo de Errores
Crea código reutilizable y a prueba de errores de usuario siguiendo estos parámetros.

1. Crea una función llamada calcular_ingreso que reciba dos parámetros: precio y cantidad.
2. Dentro de la función, utiliza un bloque try-except. Intenta multiplicar el precio por la cantidad.
3. Si ocurre un error (por ejemplo, si en lugar de un número llega un texto), el except debe atraparlo, imprimir "Error: Los valores deben ser numéricos" y retornar 0.
4. Prueba tu función enviando valores correctos y luego enviando un texto para ver cómo se maneja el error.

```python
## Haz tu codigo aquí:

def calcular_ingreso(precio, cantidad):
    try:
        ingreso = float(precio) * int(cantidad)
        return ingreso
    except:
        print("Error: Los valores deben ser numéricos")
        return 0


# Prueba con valores correctos
resultado1 = calcular_ingreso(2500.50, 3)
print("Ingreso calculado:", resultado1)

# Prueba con un valor incorrecto (texto)
resultado2 = calcular_ingreso("abc", 3)
print("Ingreso calculado:", resultado2)
```
